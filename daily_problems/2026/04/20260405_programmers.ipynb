```python
# 핵심 패턴(TWO SUM)
nums = [2, 7, 3, 6]
target = 9
d = {}

for i, n in enumerate(nums):
    if target - n in d:
        print([d[target - n], i])
    d[n] = i

print(d)
```

```python
# BFS(너비 우선 탐색) - 최단거리
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            # 아직 방문 안 한 이웃만 큐에 넣으면 중복 enqueue를 줄일 수 있음
            for nxt in graph[node]:
                if nxt not in visited:
                    queue.append(nxt)
    return visited

# DFS(깊이 우선 탐색) - 경로 탐색
# graph[node]는 보통 list. list - set 은 불가라서 for + if 로 미방문 이웃만 재귀
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    for nxt in graph[start]:
        if nxt not in visited:
            dfs(graph, nxt, visited)
    return visited
```

```python
# 내 접근
# 각 종류별로 옷을 리스트로 담은다음에 숫자를 곱하자
def solution(clothes):
    
    g = {}
    for n in clothes:
        g[n[1]].add(n[0])
    
    answer = 1
    for s in g:
        answer = answer * len[s]
        
    return answer
```

```python
def solution(n, computers):
    answer = 0
    visited = [False] * n
    
    def dfs(i):
        visited[i] = True
        
        for j in range(n):
            if computers[i][j] == 1 and not visited[j]:
                dfs(j)
                
    for i in range(n):
        if not visited[i]:
            dfs(i)
            answer +=1
    
    return answer
```
